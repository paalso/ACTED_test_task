## 2.1 Calculate the mean total income per capita and the margin of error of the mean (for the 95% confidence level) for each oblast. Present the result as a table.

```python
def moe_95(x):
    '''Сalculatу the mean error (MoE) for 95% CI'''
    n = x.count()
    if n <= 1:
        return np.nan
    std = x.std(ddof=1)
    z = 1.96  # 95% confidence level
    return z * std / np.sqrt(n)
```

```python
# Group by oblast and calculate:
income_stats_by_oblast = dfc.groupby('oblast_kiis')['income_per_capita'].agg(
    mean_income_per_capita='mean',  # average income per person in the oblast
    margin_of_error_95=moe_95       # 95% confidence interval margin of error
).reset_index()

# Sort results by oblast code
income_stats_by_oblast = income_stats_by_oblast.sort_values('oblast_kiis')

income_stats_by_oblast
```

### The livelihoods living standard gap (LSG) calculation based on income source

```python
INCOME_SOURCE_CATEGORY = {
    "idp_benefits": "assistance",
    "humanitarian_aid": "assistance",
    "loans_or_support_from_family_and_friends_within_ukraine": "assistance",
    "loans_support_or_charitable_donations_from_community_members": "assistance",

    "casual_or_daily_labour": "irregular",
    "other": "irregular",

    "salaried_work": "regular",
    "income_from_own_business_or_regular_trade": "regular",
    "income_from_own_production": "regular",
    "income_from_rent": "regular",
    "money_transfers_from_abroad_from_family_and_friends": "regular",
    "pension": "regular",

    "other_government_social_benefits_or_assistance": "assistance",

    "none": "none",      # special category → level 4
    "dont_know": "NA",
    "prefer_not_to_answer": "NA",
}
```

```python
def convert_income_source_to_lsg(record, convertion_mapping=INCOME_SOURCE_CATEGORY):
    """
    Convert income sources of a household into LSG level (1–4 or "NA").
    Based on rules from Livelihoods_LSG.xlsx.
    """
    record_income_sources_value = record["J_27_income_sources"]

    if pd.isna(record_income_sources_value):
        return "NA"

    income_sources = record_income_sources_value.split()
    categories = {convertion_mapping.get(income, "NA") for income in income_sources}

    if categories ==  { "NA" }:
        return "NA"
    
    # Income sources include only assistance OR none
    if categories in ({"assistance"}, {"none"}):
        return 4

    # Income sources includes only irreuglar income AND assistance OR Only irregular income
    if categories == { "irregular" } or categories == { "irregular", "assistance" }:
        return 3

    # Income sources are exclusively from regular income
    if categories == { "regular" }:
        return 1
    
    # Income sources include at least one regular income and some irregular or/and assistance income sources
    
    return 2
```

```python
INCOME_THRESHOLDS = {
    4: 2_324.00,
    3: 6_471.40,
    2: 9_707.10,
    1: float("inf"),   # всё, что выше → 1
}
```

```python
def convert_income_per_capita_to_lsg(record, thresholds=INCOME_THRESHOLDS):
    """
    Assign LSG level (1–4 or "NA") based on income per capita.
    Thresholds are provided externally for easy updates.
    """

    value = record["income_per_capita"]

    # NA input → NA output
    if pd.isna(value):
        return "NA"

    # Loop thresholds in order of severity (4 → 1)
    for level, limit in thresholds.items():
        if value < limit:
            return level
```

### The livelihoods living standard gap (LSG) calculation based on livelihoods coping strategies (“fsl_lcsi_XXX”)

```python
LCSI_PRIORITY_MAP = {
    'fsl_lcsi_emergency3': 4,
    'fsl_lcsi_emergency2': 4,
    'fsl_lcsi_emergency1': 4,
    'fsl_lcsi_crisis3': 3,
    'fsl_lcsi_crisis2': 3,
    'fsl_lcsi_crisis1': 3,
    'fsl_lcsi_stress4': 2,
    'fsl_lcsi_stress3': 2,
    'fsl_lcsi_stress2': 2,
    'fsl_lcsi_stress1': 2,
}
```

```python
df[reversed(LCSI_TO_LSG_MAPPING.keys())]
```

```python
def convert_lcsi_to_lsg(row, LCSI_PRIORITY_MAP):
    """
    Determines the Livelihoods LSG level for a household based on the LCSI.
    
    Rules:
    - If at least one emergency strategy is used → level 4.
    - Otherwise, if at least one crisis strategy is used → level 3.
    - Otherwise, if at least one stress strategy is used → level 2.
    - If no strategy is used → level 1.
    - If data on strategies is missing or incorrect → ‘NA’.
    """
    
    USED_STRATEGY_VALUES = {'yes', 'no_exhausted'}
    NOT_USED_VALUES = {'no_had_no_need', 'not_applicable'}

    collected_values = set()

    # стратеги уже отсортированы по приоритету в словаре
    for col, level in LCSI_PRIORITY_MAP.items():
        value = row[col]
        if value in USED_STRATEGY_VALUES:
            return level
        collected_values.add(value)

    # если явно ни одна стратегия не использовалась
    if collected_values.issubset(NOT_USED_VALUES):
        return 1

    # иначе что-то не так (пропуски/смешанные значения)
    return "NA"
```

### Create three new columns with the livelihoods living standard gap (LSG) based on the income sources, total income per capita, and livelihoods coping strategies and final Livelihoods LSG column

```python
dfc['lsg_income_sources'] = dfc.apply(
    convert_income_source_to_lsg, axis=1
)

dfc['lsg_income_per_capita'] = dfc.apply(
    convert_income_per_capita_to_lsg, axis=1
)

dfc['lsg_lcsi'] = dfc.apply(
    lambda row: convert_lcsi_to_lsg(row, LCSI_PRIORITY_MAP),
    axis=1
)
```

```python
def get_final_lsg(row):
    """
    Final Livelihoods LSG score (1–4 or "NA").

    Takes the three dimension-specific LSG values:
    - lsg_income_sources
    - lsg_income_per_capita
    - lsg_lcsi

    If at least one dimension is "NA" → return "NA".
    Otherwise → return maximum (worst) LSG level.
    """

    vals = [
        row["lsg_income_sources"],
        row["lsg_income_per_capita"],
        row["lsg_lcsi"],
    ]

    # If any dimension is NA → final LSG is NA
    if any(v == "NA" for v in vals):
        return "NA"

    # Otherwise take the maximum numeric level
    return max(vals)
```

```python
dfc['livelihoods_lsg'] = dfc[['lsg_income_sources', 
                  'lsg_income_per_capita', 
                  'lsg_lcsi']].apply(get_final_lsg, axis=1)
```
